```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
df = pd.read_csv('temperature_data.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

```python
from sklearn.linear_model import LinearRegression
# Создание необходимых колонок
import time
def create_column(df):
    start_time = time.time()
    df.sort_values(by='timestamp', inplace=True)
    df['moving_average'] = df.groupby('city')['temperature'].transform(lambda x: x.rolling(window=30).mean())
    df['moving_std'] = df.groupby('city')['temperature'].transform(lambda x: x.rolling(window=30).std())
    df['is_anomaly'] = (df['temperature'] > (df['moving_average'] + 2 * df['moving_std'])) | \
                       (df['temperature'] < (df['moving_average'] - 2 * df['moving_std']))
    season_stats = pd.DataFrame()
    season_stats['season_mean_temperature'] = df.groupby(['city', 'season'])['temperature'].mean()
    season_stats['season_std_temperature'] = df.groupby(['city', 'season'])['temperature'].std()
    season_stats = season_stats.reset_index()
    # Подсчёт тренда для каждого города с помощью линейной регрессии
    trend_results = []
    for city in df['city'].unique():
        X = np.arange(len(df[df['city'] == city])).reshape(-1, 1)
        y = df[df['city'] == city]['temperature']
        model = LinearRegression()
        model.fit(X, y)
        trend = model.coef_[0]
        trend_results.append((city, trend))

    trend_df = pd.DataFrame(trend_results, columns=['city', 'trend_value'])
    print(f"Время выполнения: {time.time() - start_time} секунд")
    df.set_index('timestamp', inplace=True)
    return df, trend_df, season_stats

df, trend_df, season_stats = create_column(df)
```

```python
# Изображение температуры и скользящего среднего на одном графике
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def plot_temperature_and_moving_average(df, city):
    df = df[df['city'] == city]
    df.index = pd.to_datetime(df.index)
    plt.figure(figsize=(15, 8))

    plt.plot(df["temperature"], label='Температура', color='steelblue')
    plt.plot(df['moving_average'], label='Скользящее среднее за 30 дней', color='orange')

    anomaly_points = df[df['is_anomaly'] == True]
    plt.scatter(anomaly_points.index, anomaly_points['temperature'], color='red', label='Аномалии', zorder=5)

    plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Разделитель по годам
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Формат отображения: только год

    plt.legend(title='', loc='upper left', fontsize=14)
    plt.xlabel('Дата', fontsize=14)
    plt.ylabel('Температура', fontsize=14)

    plt.title(f"Температура, скользящее среднее и аномалии для города {city}", fontsize=16)

    plt.show()

plot_temperature_and_moving_average(df, city='New York')
```

```python
# Расспараллелим процесс вычсления переменных, chunk = 1 city
import pandas as pd
from concurrent.futures import ProcessPoolExecutor
import time

df = pd.read_csv('temperature_data.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])

start_time = time.time()

def process_city_data(city_data):
    city_data = city_data.copy()
    city_data.sort_values(by='timestamp', inplace=True)
    city = city_data['city'].iloc[0]
    city_data['moving_average'] = city_data['temperature'].transform(lambda x: x.rolling(window=30).mean())
    city_data['moving_std'] = city_data['temperature'].transform(lambda x: x.rolling(window=30).std())
    city_data['is_anomaly'] = (city_data['temperature'] > (city_data['moving_average'] + 2 * city_data['moving_std'])) | \
                              (city_data['temperature'] < (city_data['moving_average'] - 2 * city_data['moving_std']))

    season_stats = pd.DataFrame()
    season_stats['season_mean_temperature'] = city_data.groupby('season')['temperature'].mean()
    season_stats['season_std_temperature'] = city_data.groupby('season')['temperature'].std()
    season_stats = season_stats.reset_index()
    season_stats = season_stats.assign(city=city)

    # Подсчёт тренда для каждого города с помощью линейной регрессии
    trend_results = []
    X = np.arange(len(city_data)).reshape(-1, 1)
    y = city_data['temperature']
    model = LinearRegression()
    model.fit(X, y)
    trend = model.coef_[0]
    trend_results.append((city, trend))

    trend_df = pd.DataFrame(trend_results, columns=['city', 'trend_value'])
    city_data.set_index('timestamp', inplace=True)
    return city_data, trend_df, season_stats

def create_column_parallel(df):
    cities = df['city'].unique()
    chunks = [df[df['city'] == i] for i in cities]

    with ProcessPoolExecutor() as executor:
        results = list(executor.map(process_city_data, chunks))

    city_data_list = []
    trend_data_list = []
    season_data_list = []

    for city_data, trend_df, season_stats in results:
        city_data_list.append(city_data)
        trend_data_list.append(trend_df)
        season_data_list.append(season_stats)

    combined_city_data = pd.concat(city_data_list).sort_index()
    combined_trend_data = pd.concat(trend_data_list).reset_index(drop=True)
    combined_season_data = pd.concat(season_data_list).reset_index(drop=True)

    return combined_city_data, combined_trend_data, combined_season_data

result_df = create_column_parallel(df)

result_city_data, result_trend_data, result_season_data = create_column_parallel(df)

print(f"Время выполнения: {time.time() - start_time} секунд")
```

# Мониторинг текущей температуры

```python
# Синхронный запрос через библиотеку requests
import requests
import time
import datetime

def get_current_temperature(city, api_key):
    start_time = time.time()

    # Формируем URL для запроса
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric&lang=ru'

    try:
        response = requests.get(url)

        if response.status_code == 200:
            data = response.json()
            temperature = data['main']['temp']
            timestamp = data['dt']
            current_month = datetime.datetime.utcfromtimestamp(timestamp).month

            execution_time = time.time() - start_time

            return temperature, current_month, execution_time

        else:
            execution_time = time.time() - start_time
            return None, None, execution_time

    except Exception as e:
        execution_time = time.time() - start_time
        return None, None, execution_time

api_key = ''
city = 'New York'
temperature, current_month, execution_time = get_current_temperature(city, api_key)

if temperature is not None:
    print(f'Текущая температура в городе: {temperature}°C.')
    print(f'Текущий месяц: {current_month}.')
    print(f'Время выполнения запроса: {execution_time:.2f} секунд.')
else:
    print(f'Ошибка при запросе данных. Время выполнения запроса: {execution_time:.2f} секунд.')
```

```python
# Асинхронный запрос через библиотеку aiohttp
import aiohttp
import asyncio

async def get_current_temperature_async(city, api_key):
    # Формируем URL для запроса
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric&lang=ru'

    try:
        async with aiohttp.ClientSession() as session:
            async with session.get(url) as response:
                if response.status == 200:
                    data = await response.json()
                    temperature = data['main']['temp']

                    timestamp = data['dt']
                    current_month = datetime.datetime.utcfromtimestamp(timestamp).month

                    return temperature, current_month
                else:
                    return None, None

    except Exception as e:
        return None, None

async def main(city, api_key):
    start_time = time.time()
    temperature, current_month = await get_current_temperature_async(city, api_key)
    end_time = time.time()

    execution_time = end_time - start_time

    if temperature is not None and current_month is not None:
        print(f'Текущая температура в городе {city}: {temperature}°C')
        print(f'Текущий месяц: {current_month}')
    else:
        print(f'Ошибка при запросе данных для города {city}.')

    print(f'Время выполнения запроса: {execution_time:.2f} секунд')
    return temperature, current_month

temperature, current_month = await main(city, api_key)
```

Обращение к API — это классическая I/O (input/output) задача. Ассинхронность применяется для I/O-bound задач, где большая часть времени уходит на ожидание данных от внешних систем.

Сравнивая время выполнения двух функций после нескольких заупсков можно отметить, что асинхронный подход чаще оказывается быстрее, но эта разница незначительная. В случаях, когда нужно выполнить всего один запрос, асинхронность может не дать большого выигрыша в производительности. Однако при работе с большим количеством запросов, особенно в реальном времени, асинхронный подход значительно ускоряет процесс.

```python
# Определяем сезон по месяцу
def get_current_season(current_month):
    if current_month in [1, 2, 12]:
        return 'winter'
    elif current_month in [3, 4, 5]:
        return 'spring'
    elif current_month in [6, 7, 8]:
        return 'summer'
    else:
        return 'autumn'
current_season = get_current_season(current_month)
```
